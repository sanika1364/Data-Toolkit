Q.2 How does broadcasting work in NumPy ?

 -  Broadcasting in NumPy allows arrays with different shapes to be used in arithmetic operations. It's a powerful feature that simplifies code and improves performance by avoiding the need for explicit loops.

 - When performing operations on arrays of different shapes, NumPy attempts to "broadcast" the smaller array across the larger array so that they have compatible shapes.

- This broadcasting follows a set of rules:

   - If the arrays have different numbers of dimensions, the shape of the smaller array is padded with ones on its left side.

   - If the shapes of the arrays do not match in any dimension, and neither dimension is 1, an error is raised.
  
   - If the shapes do not match in any dimension, and one of the dimensions is 1, the array with shape 1 is stretched to match the other shape.

Q.6 What are the differences between NumPy arrays and Python lists ?

 - NumPy arrays and Python lists are both used to store collections of data, but they have some key differences:

  - Performance: NumPy arrays are generally much faster for numerical operations than Python lists because they are implemented in C and use fixed-type elements, allowing for optimized vectorized operations.

  - Data Types: NumPy arrays are homogeneous, meaning all elements in an array must be of the same data type. Python lists are heterogeneous and can contain elements of different data types.

  - Functionality: NumPy arrays provide a wide range of mathematical functions and operations that can be applied directly to the entire array. Python lists have more general-purpose methods.

  - Memory Usage: NumPy arrays are more memory-efficient for storing large amounts of numerical data compared to Python lists.

  - Broadcasting: NumPy arrays support broadcasting, which allows operations on arrays with different shapes. Python lists do not have this feature.
  
  - In summary, NumPy arrays are specialized for numerical computations and offer significant performance advantages and functionality for working with numerical data, while Python lists are more versatile and can store mixed data types.

Q.8  What does the term “vectorized operation” mean in NumPy ?
  
   - A "vectorized operation" in NumPy refers to applying mathematical operations to an entire array at once, rather than iterating through each element individually using Python loops. This is a core concept in NumPy and is the reason behind its significant performance advantages for numerical computations.

  - Instead of writing explicit loops like this:

    result = []for x, y in zip(array1, array2):result.append(x + y)

  - You can perform the same operation in a vectorized way:

    result = array1 + array2

  - NumPy handles the element-wise addition internally, which is much faster because these operations are optimized and executed in compiled code (C or Fortran).

  - This vectorized approach not only makes your code more concise and readable but also dramatically improves performance, especially when dealing with large datasets.

Q.17 Explain the difference between apply() and map() in Pandas

 - In Pandas, both apply() and map() are used to apply a function to data, but they operate at different levels and have different use cases:

  - map(): This method is used on a Series to substitute each value in the Series with another value. It takes a dictionary, a Series, or a function as input. It's primarily used for element-wise transformation or mapping values based on a lookup.

  import pandas as pd

  data = {'col1': ['A', 'B', 'C', 'A']}
  
  df = pd.DataFrame(data)

  mapping = {'A': 'Apple', 'B': 'Banana', 'C': 'Cherry'}

  df['col1_mapped'] = df['col1'].map(mapping)
  
  print(df)

apply(): This method can be used on both Series and DataFrames. It applies a function along an axis of the DataFrame or Series.

When used on a Series, it behaves similarly to map(), applying a function element-wise.

When used on a DataFrame, it can apply a function to each row (axis=1) or each column (axis=0). This function can operate on multiple elements of a row or column at once.

import pandas as pd

data = {'col1': [1, 2, 3], 'col2': [4, 5, 6]}

df = pd.DataFrame(data)

df['col1_plus_col2'] = df.apply(lambda row: row['col1'] + row['col2'], axis=1)

print(df)

 - Key Differences:

    - Scope: map() is strictly for Series, while apply() works on both Series and DataFrames.

    - Functionality: map() is primarily for element-wise value substitution or mapping. apply() is more general and can apply functions that operate on multiple elements (rows or columns) at a time when used on a DataFrame.

    - Performance: For simple element-wise operations on a Series, map() is often faster than apply() because it is optimized for this task. However, for more complex operations or operations on DataFrames, apply() is necessary.
  - In summary, use map() for simple element-wise transformations on a Series, and use apply() for more complex operations on Series or for applying functions along rows or columns of a DataFrame.

Q.21  Why is NumPy’s array slicing faster than Python’s list slicing ?

 - NumPy array slicing is generally faster than Python list slicing due to several underlying reasons related to how they are implemented and stored in memory:

    - Contiguous Memory Allocation: NumPy arrays are typically stored in a contiguous block of memory. This allows for efficient access to elements and optimized operations, including slicing. Python lists, on the other hand, are implemented as arrays of pointers to objects, which can be scattered across memory.

    - Fixed Data Type: NumPy arrays are homogeneous, meaning all elements are of the same data type. This allows NumPy to work directly with the raw data in memory without the overhead of checking the type of each element during operations like slicing. Python lists can contain elements of different types, requiring more overhead during access and slicing.

    - Optimized Implementation: NumPy slicing operations are implemented in C, which is a lower-level language than Python. This allows for highly optimized and efficient code execution compared to the more general-purpose implementation of Python list slicing.

    - Vectorized Operations: NumPy's vectorized operations, which are often used in conjunction with slicing, further contribute to performance gains. These operations are applied to entire arrays or slices at once, leveraging optimized low-level code.

 - In summary, the contiguous memory layout, fixed data type, and optimized C implementation of NumPy arrays contribute significantly to the performance advantage of NumPy array slicing over Python list slicing, especially when dealing with large datasets.
